```python
import re
from datetime import datetime, timedelta

def convert_relative_date(relative_date):
    today = datetime.now()
    
    if 'now' in relative_date.lower():
        return today.strftime('%Y/%m/%d')
    
    match = re.search(r'(\d+)', relative_date)
    if match:
        days_ago = int(match.group(1))
        published_date = today - timedelta(days=days_ago)
        return published_date.strftime('%Y/%m/%d')
    
    return None  # Or handle other cases if needed

# Example usage:
relative_dates = ["Posted 1 day ago", "Posted 4 days ago", "Posted now"]
converted_dates = [convert_relative_date(date) for date in relative_dates]

for original, converted in zip(relative_dates, converted_dates):
    print(f"{original} -> {converted}")
```
