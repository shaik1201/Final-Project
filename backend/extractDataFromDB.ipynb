```python
from flask import Flask, jsonify, request
from flask_cors import CORS
from pymongo import MongoClient
import pandas as pd
import config

# Initialize Flask app
app = Flask(__name__)
CORS(app)

# Database setup
client = MongoClient(config.MONGO_URI)
db = client['job_search_db']
collection = db['jobs']  # Ensure this is the correct collection name

# Debugging: Check database and collection
print("Database:", db)
print("Collection:", collection)

# Fetch all documents from the collection
cursor = collection.find()
print("Cursor:", cursor)

# Convert the cursor to a list and then to a DataFrame
data = list(cursor)
if not data:
    print("No data found in the collection.")
else:
    df = pd.DataFrame(data)
    print("DataFrame:\n", df)

    # Optionally, remove the MongoDB _id field if not needed
    if '_id' in df.columns:
        df.drop('_id', axis=1, inplace=True)

    # Save the DataFrame to a CSV file
    df.to_csv('output.csv', index=False)
    print("Data has been exported to output.csv")
```
